```python
# Создание DataFrame
df = pd.DataFrame(data, columns=['A', 'B', 'C'])

# Построение попарных диаграмм рассеяния
sns.pairplot(df)
plt.show()

# Вычисление коэффициентов корреляции
correlation_matrix = df.corr()
print("Коэффициенты корреляции:")
print(correlation_matrix)

# Проверка значимости коэффициентов корреляции
def correlation_significance_test(df):
    p_values = pd.DataFrame(columns=df.columns, index=df.columns)
    for row in df.columns:
        for col in df.columns:
            if row == col:
                p_values.loc[row, col] = np.nan
            else:
                _, p_value = stats.pearsonr(df[row], df[col])
                p_values.loc[row, col] = p_value
    return p_values

p_values = correlation_significance_test(df)
print("\nP-значения для проверки значимости коэффициентов корреляции:")
print(p_values)

# Вывод о зависимости величин
def interpret_correlation(correlation_matrix, p_values, alpha=0.05):
    interpretation = {}
    for row in correlation_matrix.index:
        for col in correlation_matrix.columns:
            if row != col:
                corr = correlation_matrix.loc[row, col]
                p_value = p_values.loc[row, col]
                if p_value < alpha:
                    if corr > 0:
                        relation = "положительная корреляция"
                    else:
                        relation = "отрицательная корреляция"
                    interpretation[(row, col)] = f"Значимая {relation} (r = {corr:.2f}, p = {p_value:.2g})"
                else:
                    interpretation[(row, col)] = f"Не значимая корреляция (r = {corr:.2f}, p = {p_value:.2g})"
    return interpretation

interpretation = interpret_correlation(correlation_matrix, p_values)
for key, value in interpretation.items():
    print(f"{key}: {value}")
```

### Task 3.1 Гипотезы

```python
# Данные
A = np.array([11.11, 10.94, 11.03, 10.97, 10.94, 10.87, 11.01, 10.88, 11.07, 10.92, 10.98, 11.1, 10.85, 11.18, 10.87])
B = np.array([11.08, 10.92, 10.96, 11.08, 10.98, 10.86, 11.02, 11.12, 10.94, 10.75, 11.01, 10.9, 10.99, 11.01, 10.93])
```

```python
# Проверка равенства средних (t-тест Стьюдента для независимых выборок)
t_statistic, p_value_mean = stats.ttest_ind(A, B)

# Проверка равенства дисперсий (тест Фишера)
f_statistic, p_value_var = stats.levene(A, B)

print(f"Результаты t-теста для проверки равенства средних:")
print(f"t-статистика = {t_statistic:.4f}, p-значение = {p_value_mean:.4f}")

print(f"\nРезультаты теста Фишера для проверки равенства дисперсий:")
print(f"F-статистика = {f_statistic:.4f}, p-значение = {p_value_var:.4f}")

# Интерпретация результатов
alpha = 0.05
if p_value_mean < alpha:
    print("\nСредние значения значимо различаются.")
else:
    print("\nСредние значения не различаются значимо.")

if p_value_var < alpha:
    print("Дисперсии значимо различаются.")
else:
    print("Дисперсии не различаются значимо.")
```
